```r
Lake <-c('Loch Ness', 'Loch Lomond',  'Loch Morar',   'Loch Tay', 'Loch Awe', 'Loch Maree', 'Loch Ericht', 'Loch Lochy', 'Loch Rannoch', 'Loch Shiel',  'Loch Katrine', 'Loch Arkaig',  'Loch Shin') 
Volume <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
Area <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
Lake_Length  <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
Max_Depth <- c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49)
Mean_Depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)

scottish.lakes <- data.frame(Lake, Volume, Area, Lake_Length, Max_Depth, Mean_Depth)
scottish.lakes
V_max_i <- which.max(scottish.lakes$Volume)
V_min_i <- which.min(scottish.lakes$Volume)
A_max_i <- which.max(scottish.lakes$Area)
A_min_i <- which.min(scottish.lakes$Area)
print(paste("Largest Volume Lake", scottish.lakes[V_max_i,1], ", Volume[km^3] = ", scottish.lakes[V_max_i,2] ))
print(paste("Smallest Volume Lake", scottish.lakes[V_min_i,1], ", Volume[km^3] = ", scottish.lakes[V_min_i,2] ))
print(paste("Largest Area", scottish.lakes[A_max_i,1], ", Area[km^2] = ", scottish.lakes[A_max_i,3] ))
print(paste("Smallest Area", scottish.lakes[A_min_i,1], ", Area[km^2] = ", scottish.lakes[A_min_i,3] ))
```

```r
#install.packages("dplyr")
library(dplyr)
coal_prod<- read.csv("coal-production-by-country.csv")
coal_tbl <- as_data_frame(coal_prod)
class(coal_tbl) # tibble verification
colnames(coal_tbl)[1] <- "Country"
colnames(coal_tbl)[4] <- "coal_prod_TWh"
#coal_tbl
print(coal_tbl)
countries <-unique(coal_tbl["Country"])
c_entry <-table(coal_tbl["Country"]) 
```

```r
countries['frequency']<-as.numeric(c_entry)
print(countries)
```

```r
library("ggplot2")
mask<- countries["frequency"]>40
c_mask <- countries[mask,]
p<-ggplot(data=c_mask, aes(x=Country, y=frequency)) +
  geom_bar(stat="identity", fill="lightblue")
p + theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +scale_fill_hue(l=40, c=35)+ 
ylab("Entries") +
ggtitle("World Coal production for countries >40  entries")
```

```r
yr_mask<- coal_tbl["Year"]>1970
r_coal_tbl<-coal_tbl[yr_mask,]
print(r_coal_tbl)
```

```r
i_prod<-aggregate(r_coal_tbl$"coal_prod_TWh",list(r_coal_tbl$Country), sum)
colnames(i_prod)[1] <- "Country"
colnames(i_prod)[2] <- "Total Integrated Coal Production"
#i_prod
i_prod<-i_prod[order(-i_prod$"Total Integrated Coal Production"),]
i_prod
```

```r
top5<-head(i_prod, n=5)
top5
```

```r
v<-c(top5$Country)
print(top5$Country[1])
 
mask_c1 <- r_coal_tbl[1]==toString(top5$Country[1]) #
mask_c2 <- r_coal_tbl[1]==toString(top5$Country[2])
mask_c3 <- r_coal_tbl[1]==toString(top5$Country[3])
mask_c4 <- r_coal_tbl[1]==toString(top5$Country[4])
mask_c5 <- r_coal_tbl[1]==toString(top5$Country[5])

t5_coal<-r_coal_tbl[mask_c1 | mask_c2 | mask_c3 | mask_c4 | mask_c5,]
#t5_coal 

ggplot(data=t5_coal, aes(x=Year, y=coal_prod_TWh ,group=Country, colour = Country))  + geom_line() +
ylab(" Coal production (TWh)")
```

```r
vacc_manu<- read.csv("vaccinations-by-manufacturer.csv")
vacc_manu_tbl <- as_data_frame(vacc_manu)
print(vacc_manu_tbl)
```

```r
italy<-filter(vacc_manu_tbl, vacc_manu_tbl[1]=="Italy", .preserve = FALSE)
print(italy)
```

```r
p_it<- ggplot(data=italy, aes(x=date, y=total_vaccinations, group = vaccine, color=vaccine))
p_it + 
geom_line()+ 
theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +scale_fill_hue(l=40, c=35)+
scale_x_discrete(breaks = levels(italy$date)[floor(seq(1, nlevels(italy$date), length.out = 10))])+
ylab("Total vaccinations")+ ggtitle("Italy vaccinations by manufacturer")

it_sum<-italy %>%
    group_by(date) %>%
     summarise(Sum = sum(total_vaccinations))
print(it_sum)
p_it_sum<- ggplot(data=it_sum, aes(x=date, y=Sum)) 
p_it_sum + 
geom_point(colour="blue") + 
theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))+
scale_x_discrete(breaks = levels(it_sum$date)[floor(seq(1, nlevels(it_sum$date), length.out = 10))])+
ylab("Total Vaccinations") + ggtitle("Italy Vaccinations (All manufacturers)")
```
